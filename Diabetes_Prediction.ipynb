```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import seaborn as sns

from sklearn.model_selection import train_test_split

#import dataset
df = pd.read_csv('diabetes.csv')

#Encode Smoking_history variable
df['smoking_history'] = df['smoking_history'].replace('never', 1)
df['smoking_history'] = df['smoking_history'].replace('not current', 2)
df['smoking_history'] = df['smoking_history'].replace('former', 3)
df['smoking_history'] = df['smoking_history'].replace('current',4)
df['smoking_history'] = df['smoking_history'].replace('ever', 5)
df['smoking_history'] = df['smoking_history'].replace('No Info', 6)
```

```python
# Histogram Distribution of Age
x = df['age']
fig, axs = plt.subplots(1, 1, figsize=(10, 7), tight_layout=True)

for s in ['top', 'bottom', 'left', 'right']:
    axs.spines[s].set_visible(False)

axs.xaxis.set_ticks_position('none')
axs.yaxis.set_ticks_position('none')
axs.xaxis.set_tick_params(pad=5)
axs.yaxis.set_tick_params(pad=10)
axs.grid(True, color='black', linestyle='-.', linewidth=0.5, alpha=0.6)

N, bins, patches = axs.hist(x, bins=30)
fracs = ((N**(1 / 5)) / N.max())
norm = colors.Normalize(fracs.min(), fracs.max())

for thisfrac, thispatch in zip(fracs, patches):
    color = plt.cm.viridis(norm(thisfrac))
    thispatch.set_facecolor(color)

plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title('Distribution of Age')
plt.show()

# PieChart of Gender Distribution
gen_counts = df['gender'].value_counts()
labels = ['Female', 'Male', 'Other']
myexplode = [0.2, 0, 0]
colors = ['tomato', 'royalblue', 'olive']
plt.figure(figsize=(8, 8))
plt.pie(gen_counts, labels=labels, colors=colors, explode=myexplode, shadow=True, autopct='%1.1f%%')
plt.legend(labels, loc='upper right')
plt.title('Gender Distribution')
plt.show()

# Correlation Heatmap
df_updated = df.drop(columns=['gender'])
plt.figure(figsize=(14, 8))
heatmap = sns.heatmap(df_updated.corr(), cmap='coolwarm', vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':14}, pad=12)
plt.show()
```
